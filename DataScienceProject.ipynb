# Section 4: Data Science 

Question: Basketball fans often said that LeBron James is a great and selfless player to make his teammates score better. Is the statement true?
To examine this I use LeBron's assist percentage and assist data throughout his carrer to examine the question above.A player's assist percentage shows how often the player is assisting teammates and represents the proportion of team scoring for which a player is responsible through assists.To see how correlated is assistant percentage and assists for Lebron James, I use the .corr() function for his data set. Then, I also illustrate the scatter plot to visualize and the trend between assist percentage and assist.

```python
#plot the scatter plot between assist percentage and assist and find the linear relationship
import seaborn as sns
sns.lmplot(x="ast_pct",y="ast",data=df_lebron)
#the correlation between assist percentage and assist of LeBron James throughout his 20 years carrer
df_lerbon_corr=df_lebron.select_dtypes(include="number").corr()
df_lerbon_corr["ast_pct"]["ast"]



```

Not only that,we can use assist percentage of LeBron James to predict his amount of assists without knowing them by putting them into a model of liner regression. We can see the assist predicted in the dataset below.

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
ind=["ast_pct"]
dep="ast"
model.fit(df_lebron[ind],df_lebron[dep])
df_lebron["ast_predict"]=model.predict(df_lebron[ind])
df_lebron
```

We can find prediction error by subtracting the actual assist and the predicted assist.

```python
df_lebron["error"]=df_lebron["ast"]-df_lebron["ast_predict"]
error=df_lebron["error"].sum()
error
```
